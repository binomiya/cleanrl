```python
#ある軸について90度回転する行列
rotation_x = np.array([[1,0,0], [0, 0, -1], [0, 1, 0]])
rotation_z = np.array([[0,1,0], [-1, 0, -0], [0, 0, 1]])
```

```python
kaiten = np.array([[0,-1,0],[-1,0,0],[0,0,1]])
#x軸についての90度回転版を作成
x_rotated_x90 = []
y_rotated_x90 = []
z_rotated_x90 = []

for i in range(120):

    X_rotated_x90 = []
    Y_rotated_x90 = []
    Z_rotated_x90 = []

    for j in range(22):
        
        point_pre_rotation = np.array([dataDict_forplot[i][j][0], dataDict_forplot[i][j][1], dataDict_forplot[i][j][2]])
        point_post_rotation = np.matmul(rotation_x, point_pre_rotation)
        point_post_post_rotation = np.matmul(kaiten, point_post_rotation)
        
        X_rotated_x90.append(point_post_post_rotation[0])
        Y_rotated_x90.append(point_post_post_rotation[1])
        Z_rotated_x90.append(point_post_post_rotation[2])

    x_rotated_x90.append(X_rotated_x90)
    y_rotated_x90.append(Y_rotated_x90)
    z_rotated_x90.append(Z_rotated_x90)

#x[flame][joint][]
```

```python
#x軸について90度回転し、z軸について90度回転した座標を作成
x_rotated_x90_z90 = []
y_rotated_x90_z90 = []
z_rotated_x90_z90 = []

for i in range(120):

    X_rotated_x90_z90 = []
    Y_rotated_x90_z90 = []
    Z_rotated_x90_z90 = []

    for j in range(22):
        
        point_pre_rotation = np.array([dataDict_forplot[i][j][0], dataDict_forplot[i][j][1], dataDict_forplot[i][j][2]])
        point_post_rotation_x = np.matmul(rotation_x, point_pre_rotation)
        point_post_rotation_xz = np.matmul(rotation_z, point_post_rotation_x)
        
        X_rotated_x90_z90.append(point_post_rotation_xz[0])
        Y_rotated_x90_z90.append(point_post_rotation_xz[1])
        Z_rotated_x90_z90.append(point_post_rotation_xz[2])

    x_rotated_x90_z90.append(X_rotated_x90_z90)
    y_rotated_x90_z90.append(Y_rotated_x90_z90)
    z_rotated_x90_z90.append(Z_rotated_x90_z90)

#x[flame][joint][]
```
